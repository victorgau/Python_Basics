```python
# 載入需要的模組
import geocoder
import folium
```

```python
# 使用 geocoder 取得特定住址的 GPS 座標
location = geocoder.osm('國立高雄師範大學').latlng
```

```python
# 顯示 GPS 座標
location
```

```python
m = folium.Map(location=location, zoom_start=16)
```

```python
folium.Marker(location=location, popup='國立高雄師範大學').add_to(m)
```
